preprocessing the data.

```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels= []
train_samples = []
```

```python
for i in range(1000):
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(0)

  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(1)

for i in range(50):
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(1)

  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(0)
```

```python
#print raw data

for i in train_samples:
  print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
for i in scaled_train_samples:
  print(i)
```

Learning

```python
import keras
from keras import backend as k
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Activation
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model1 = Sequential([
    Dense(16, input_shape= (1,), activation= 'relu'),
    Dense(32, activation='relu'),
    Dense(2,activation= 'softmax')
])
```

```python
model1.compile(Adam(learning_rate =.0001), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model1.fit(scaled_train_samples, train_labels, validation_split = 0.2 , batch_size=10, epochs=20, shuffle=True, verbose=2)
```

####  A single pass of data through the model is called an epoch, After a single epoch same data will be passed over and over through multiple epoch and during this repeated process the model will be learning

Batch size is how many pieces data we want to send at once
